Gauss_logit_SB_VAE_RESULTS

```python
import pandas as pd
```

```python
df = pd.read_table('gauss_logit_SB_VAE_results_.txt',sep="\t",header=None)
```

```python
df.head()
```

```python
f=open('gauss_logit_SB_VAE_results_.txt',"r")
lines=f.readlines()
results=[]
for x in lines:
    results.append(x.split(" ")[5:6])
    for w in results:
        s='\t'.join(w)
        #print(s) 
```

```python
gauss_logit = pd.DataFrame(results,columns=['Reconstruction_Error'])
gauss_logit.tail()
```

```python
gauss_logit = gauss_logit.drop(353, axis=0)
gauss_logit.tail()
```

```python
gauss_logit.info()
```

```python
gauss_logit['Reconstruction_Error'] = gauss_logit['Reconstruction_Error'].replace(',','', regex=True)
gauss_logit.tail()
```

```python
gauss_logit['Reconstruction_Error'] = gauss_logit['Reconstruction_Error'].astype('str').astype('float')
```

```python
import matplotlib.pyplot as plt
x = [x for x in range(0,353)]
y = gauss_logit['Reconstruction_Error']
plt.figure(figsize=(8,6))
plt.xlabel('Training Epochs')
plt.ylabel('Reconstruction_Error')
plt.plot(x,y)
plt.legend('Gauss_logit_SB_VAE')
```
